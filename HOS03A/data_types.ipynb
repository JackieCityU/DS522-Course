```python
import numpy as np
```

```python
# create structured data type
dt = np.dtype([('name', 'U20'), ('age', np.int64), ('points', np.float64)])
print(dt)

# <U20 is a Unicode string with a maximum length of 20 characters
# <i8 is a 64-bit integer
# <f8 is a 64-bit float

# apply it to ndarray object called student
student = np.array([('student_A', 20, 100),
                    ('student_B_has_a_very_long_name', 20, 98),
                    ('student_C', 21, 95)], dtype=dt)

student
```

```python
# To access content of age column
student['age']
```

```python
#To access content of name column
student['name']
```

```python
#To access content of points column
student['points']
```

Challenge answers:
Challenge: What are the primary use cases 
for multi-dimensional arrays, and why does NumPy’s 
contiguous memory 
allocation outperform Python's nested lists?"
Primary use cases of multi-dimensional array are data analysis, machine learning
image processing, scientific computing where data is structured in grids or matrices.

Why NumPy is faster:
NumPy arrays use contiguous memory allocation, allowing 
efficient CPU caching, vectorization, and execution 
in C code, whereas Python nested lists store elements as separate objects 
with pointers, leading to slower performance.
